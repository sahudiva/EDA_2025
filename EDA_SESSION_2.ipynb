**iloc-loc**
- we have a drawback with take
- take method will give either rows or columns at a time
- But iloc and loc will give both rows and columns at a time

**syntax**
- dataframename.iloc[rows,columns]
- dataframename.iloc[start:stop:step,start:stop:step]

```python
visa_df.iloc[10:15,3:7]
```

```python
visa_df.iloc[[100,200,300],[2,6,8]]
```

```python
visa_df.iloc[100:105,[2,6,8]]
```
